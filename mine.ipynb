```python
import pandas as pd
Series_A = pd. Series ([1, 2, 3, 4, 5])
Series_B = pd. Series ([2, 4, 6, 8, 10])
not_common = pd. Series (list(set(Series_A) ^ set (Series_B)))
print("Q (Not common to both):")
print(not_common.tolist())
```

```python
import pandas as pd

Series_A = pd. Series ([1, 2, 3, 4, 5])

Series_B = pd. Series ([2, 4, 6, 8, 10])

# *Items common to both (Intersection, p):*

common = pd.Series(list(set (Series_A) & set (Series_B)))

print("p (Common to both):")

print(common.tolist())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("/content/employee1.CSV")
dept_salary = df.groupby("Department")["Salary"].mean()
dept_salary.plot(kind="bar", color="pink",edgecolor="black")
plt.title("Average Salary by Department")
plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("/content/employee1.CSV")
plt.hist(df['Bonus %'], bins=8, color='lightgreen',edgecolor='black', alpha=0.7)
plt.title("Bonus % Distribution of Employees")
plt.xlabel("Bonus % Amount ($)")
plt.ylabel("Number of Employees")
plt.grid(axis="y")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("/content/employees (1).csv")
male = df[df['Gender'] == 'Male']
plt.scatter(male['Salary'], male['Bonus %'], color='blue', label='Male', alpha=0.7)
female = df[df['Gender'] =='Female']
plt.scatter(female['Salary'], female['Bonus %'], color='pink', label='Female', alpha=0.7)
plt.title("Scatter Plot: Salary vs Bonus by Gender")
plt.xlabel("Salary")
plt.ylabel("Bonus %	")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("/content/employees (1).csv")
senior_df = df[df["Senior Management"] == True]
plt.figure(figsize=(6,5))
plt.boxplot(senior_df["Salary"], vert=True, patch_artist=True)
plt.title("Salary Distribution - Senior Management")
plt.ylabel("Salary")
plt.grid(True)
plt.show()
```
