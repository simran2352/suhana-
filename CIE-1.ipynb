```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/mtcars (1).csv')
auto = (df['am'] == 0).sum()
manual = (df['am'] == 1).sum()
plt.bar(['Automatic', 'Manual'], [auto, manual], color=['orange', 'blue'])
plt.title("Frequency of Transmission Types")
plt.ylabel("Number of Cars")
plt.grid(axis='y')
plt.show()
```
