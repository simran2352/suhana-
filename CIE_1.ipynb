```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/mtcars 2.csv')
auto_count=(df['am']==0).sum()
manual_count=(df['am']==1).sum()
labels=['automatic,manual']
counts=['auto_count,manual_count']
plt.bar(labels,counts,color=['pink','black'])
plt.title("frequency distribution transmission type")
plt.xlabel('distrubation')
plt.ylabel('number of cars')
plt.show()
```
